```python
# Библиотеки для работы с данными
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

import warnings
warnings.filterwarnings("ignore")

# Библиотеки для моделирования
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Оценка моделей
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score, roc_curve
```

<h1  style="text-align: center" >2. Load Data</h1><a id = "2"></a>

<h1>About Dataset</h1>
<h2>Context </h2>
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

<h2>Content</h2>
The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

<h2>Acknowledgements</h2>
Smith, J.W., Everhart, J.E., Dickson, W.C., Knowler, W.C., & Johannes, R.S. (1988). Using the ADAP learning algorithm to forecast the onset of diabetes mellitus. In Proceedings of the Symposium on Computer Applications and Medical Care (pp. 261--265). IEEE Computer Society Press.

```python
def del_outliers(data_col, df):
    """
    Функция для удаления выбросов, выходящих за пределы (среднее +- 3*с.к.о.)
    data_col - столбец данных
    df - dataframe
    """
    mean, std = df[data_col].mean(), df[data_col].std()
    cut = std * 3
    lower, upper = mean - cut, mean + cut
    new_df = df[(df[data_col] < upper) & (df[data_col] > lower)]
    return new_df
```

```python
#заменяем нулевые значения на средние или медианные
df['Insulin'] = df.Insulin.replace(0, df.Insulin.median())
df['BloodPressure'] = df.BloodPressure.replace(0,df.BloodPressure.median())
df['BMI'] = df.BMI.replace(0,df.BMI.mean())
```

```python
#преобразуем некоторые столбцы в новые (с удалением выбросов)
df  = del_outliers('Pregnancies',df)
df  = del_outliers('Glucose',df)
df  = del_outliers('BloodPressure',df)
df  = del_outliers('SkinThickness',df)
df  = del_outliers('Insulin',df)
df  = del_outliers('DiabetesPedigreeFunction',df)
df  = del_outliers('Age',df)
```

```python
df.var()
```

```python
#удаляем DPF в связи с очень маленькой дисперсией 
df=df.drop('DiabetesPedigreeFunction',axis=1)
```

<h1 style="text-align: center" >5. Data Preprocessing </h1><a id = "5"></a>

```python
#формируем зависимые переменные x и независимую переменную Y
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
col = df.select_dtypes(include=np.number).columns
col
```

```python
#преобразование через метод StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
X = pd.DataFrame(X, columns=col[:7])
```

<h1 style="text-align: center" >6. Models </h1><a id="6"></a>

```python
#Разделение данных на тестовую и тренировочную подвыборки
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
x_train.shape, x_test.shape
```

```python
#применение метод "случайные леса"
rd = RandomForestClassifier(random_state=0)
rd.fit(x_train, y_train)
```

```python
#оценивание модели с помощью метрики accuracy
y_pred = rd.predict(x_test)
accuracy_score(y_pred, y_test)
```

```python
%%time
#расчет наилучших параметров для улучшения качества модели
parameters = {"n_estimators":[10,25,50],
             "criterion": ["gini","entropy"],
             "max_depth": [4, 5],
             "max_features": ["sqrt", "log2"],
             "min_samples_split": [4,7]
             }

rsc = GridSearchCV(estimator=RandomForestClassifier(),
                  param_grid = parameters,
                  cv = 3,
                  verbose=1,
                  scoring='accuracy',
                  n_jobs=1)
rsc.fit(x_train, y_train)

print("Best parameters : ",rsc.best_params_)
print("Best scor {:.3f} ".format(rsc.best_score_))
```

```python
#Случайные леса при наилучших параметрах
rd2 = RandomForestClassifier(**rsc.best_params_)
rd2.fit(x_train, y_train)
```

```python
#матрица ошибок
cnf_matrix = confusion_matrix(rd2.predict(x_test), y_test)

plt.figure(figsize=(12,12))
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
roc_auc_score(y_test,rd2.predict(x_test))
```

```python
#оценка ROC-AUC-Curve
fpr, tpr, thresholds = roc_curve(y_test, rd2.predict(x_test))

plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='RFC')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC curve')
plt.show()
```

# Conclusion

```python
conclusion = {
    "Score": {
    "Baseline Model ": accuracy_score(rd.predict(x_test), y_test)*100,
    "Model after hyperparameter tuning": rsc.best_score_*100
    }
}

ax = pd.DataFrame(conclusion).plot(kind="bar", 
                             figsize = (10,5), 
                             legend =False, 
                             title = "Score",
                             color = 'y');
                    
for p in ax.patches:
    ax.annotate(str(round(p.get_height(), 1)), (p.get_x() * 1.005, p.get_height() * 1.005))
```
